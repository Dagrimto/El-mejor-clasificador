# Pre procesamiento: selección/extracción de características


### Veamos el día de la semana en que la gente obtuvo el préstamo


```python
df['dayofweek'] = df['effective_date'].dt.dayofweek
bins = np.linspace(df.dayofweek.min(), df.dayofweek.max(), 10)
g = sns.FacetGrid(df, col="Gender", hue="loan_status", palette="Set1", col_wrap=2)
g.map(plt.hist, 'dayofweek', bins=bins, ec="k")
g.axes[-1].legend()
plt.show()
```

Observamos que la gente que obtuvo el préstamo al final de la semana no lo pagó. Utilicemos la binarización para establecer unops valores límite menores de 4


```python
df['weekend'] = df['dayofweek'].apply(lambda x: 1 if (x>3)  else 0)
df.head()
```

## Convertir características categóricas en valores numéricos


Observemos la característica de genero:


```python
df.groupby(['Gender'])['loan_status'].value_counts(normalize=True)
```

86% de las mujeres pagan sus prestamos mientras que el 73% de los hombres lo hacen.


Convirtamos en 0 a los hombres y a las mujeres en 1:


```python
df['Gender'].replace(to_replace=['male','female'], value=[0,1],inplace=True)
df.head()
```

#### ¿Qué hay acerca de la educación?

#### Característica antes de la codificación One Hot


```python
df[['Principal','terms','age','Gender','education']].head()
```

#### Utilice la técnica de codificación One Hot para transformar las variables categóricas en binarias y agrégelas al dataframe de características


```python
Feature = df[['Principal','terms','age','Gender','weekend']]
Feature = pd.concat([Feature,pd.get_dummies(df['education'])], axis=1)
Feature.drop(['Master or Above'], axis = 1,inplace=True)
Feature.head()
```

```python
X = Feature
X[0:5]
```

# K Nearest Neighbor(KNN)

Observación: Deberá encontrar el mejor valor de k para construir el modelo con la mejor eficacia.
**Advertencia:** No deberá utilizar **loan_test.csv** para encontrar el mejor valor de k, sin embargo, puede dividir train_loan.csv para entrenamiento y pruebas para encontrar el mejor valor de **k**.


```python
from sklearn.model_selection import train_test_split
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.2, random_state=4)
print('Conjunto de entrenamiento:', X_entrenamiento.shape, y_entrenamiento.shape)
print('Conjunto de prueba:', X_prueba.shape, y_prueba.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 6

neighK6 = KNeighborsClassifier(n_neighbors=k).fit(X_entrenamiento, y_entrenamiento)
y_pred = neighK6.predict(X_prueba)

from sklearn import metrics
print("Precisión del conjunto de entrenamiento: ", metrics.accuracy_score(y_entrenamiento, neighK6.predict(X_entrenamiento)))
print("Precisión del conjunto de prueba: ", metrics.accuracy_score(y_prueba, y_pred))
```

```python
Ks = 10
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

ConfustionMx = []
for n in range(1, Ks):
    neigh = KNeighborsClassifier(n_neighbors=n).fit(X_entrenamiento, y_entrenamiento)
    y_pred = neigh.predict(X_prueba)
    mean_acc[n-1] = metrics.accuracy_score(y_prueba, y_pred)
    std_acc[n-1] = np.std(y_pred==y_prueba) / np.sqrt(y_pred.shape[0])

plt.plot(range(1, Ks), mean_acc, 'g')
plt.fill_between(range(1, Ks), mean_acc - 1 * std_acc, mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Precisión', '+/- 3xstd'))
plt.ylabel('Precisión')
plt.xlabel('Número de Vecinos (K)')
plt.tight_layout()
plt.show()
print("Mejor precisión:", mean_acc.max(), "k=", mean_acc.argmax()+1)
```

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, f1_score, jaccard_score

def plot_confusion_matrix(cm, classes, normalize=False, title='Matriz de confusión', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('Etiqueta real')
    plt.xlabel('Etiqueta predicha')

cnf_matrix = confusion_matrix(y_prueba, yhat, labels=['PAIDOFF','COLLECTION'])
np.set_printoptions(precision=2)

print(classification_report(y_prueba, yhat))

plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['PAIDOFF','COLLECTION'], normalize=False, title='Matriz de confusión')

f1 = f1_score(y_prueba, yhat, average='weighted')
jaccard = jaccard_score(y_prueba, yhat, pos_label='PAIDOFF')

print("Puntaje F1:", f1)
print("Puntaje Jaccard:", jaccard)
```
